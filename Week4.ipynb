## Text data compression system

```python
try:
    paragraph =input("paragraph:")
    if not isinstance(paragraph, str):
            raise AttributeError("Input must be a string.")
    data={}
    words = paragraph.split()
    for i in words:
        i=i.lower()
        if i in data:
            data[i]+=1
        else:
            data[i]=1
    data_tuple_list=[(word , freq) for word,freq in data.items()]
    sorted_list = sorted(data_tuple_list, key=lambda x: x[1])
    print("\nWord frequency sorted by descending order:")
    for word, freq in sorted_list:
        print(f"{word}: {freq}")
    while True:
        try:
            key_=input("Enter word to check Frequency:")
            if key_=="exit":
                break
            if key_ in data:
                print(f"The frequency of '{key_}' is: {data[key_]}")
            else:
                raise KeyError(f"Word '{key_}' not found in the text.")
            
        except Exception as e:
            print(e)

except AttributeError as e:
    print(e)
except Exception as e:
    print(e)
```

## Employee monthly payroll

```python
class NegativeSalaryError(Exception):
    def __init__(self, message):
        super().__init__(message)
        self.message = message
    def __str__(self):
        return f"{self.message})"

def generate_salary_slip(employee_tuple, employee_id,change):
    name, base_salary, department = employee_tuple
    
    if change<0:
        bonus = 0
        deductions = -change
    else:
        bonus=change
        deductions=0  
    final_salary = base_salary + bonus - deductions


    salary_slip = f"""
    ===========================
    SALARY SLIP
    ===========================
    Employee ID: {employee_id}
    Name: {name}
    Department: {department}
    ---------------------------
    Base Salary: ${base_salary}
    Bonus: +${bonus}
    Deductions: -${deductions}
    ---------------------------
    Final Salary: ${final_salary}
    ===========================
    """
    
    return salary_slip

employee_record={}
while True:
    try:
        print("1.Add Employee \n2.Search Employee & Update \n3.List Employee \n4.exit")
        opt=int(input("Enter Option:"))
        if opt==4:
            break
        elif (opt>4 or opt<1):
            print("choose option from 1-3")
            continue
        elif (opt==1):
            employee_id=input("Employee id:")
            name=input("Name:")
            base_salary=int(input("Base Salary:"))
            if base_salary < 0:
                raise NegativeSalaryError("Negative salary error")
            department=input("Department:")
            employee_record[employee_id]=(name,base_salary,department)
        elif (opt==2):
            employee_id=input("Employee id:")
            if employee_id in employee_record:
                print(f"{employee_id}:{employee_record[employee_id]}")
            else:
                raise KeyError(f"'{employee_id}' not found in the record.")
            
            ques=input("Enter yes to update salary:")

            if ques.lower()=="yes":
                change=int(input("Enter value to adjust use -ve for deduction"))
                if (employee_record[employee_id][1]+change<0):
                    raise NegativeSalaryError("Negative salary error")
                else:
                    print(generate_salary_slip(employee_record[employee_id],employee_id,change))
                    employee_record[employee_id]=(employee_record[employee_id][0],employee_record[employee_id][1]+change,employee_record[employee_id][2])
        elif (opt==3):
            for key, values in employee_record.items():
                print(f"{key}:{values}")

    except Exception as e:
        print(e)
```
